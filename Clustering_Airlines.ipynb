```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
data=pd.read_csv("D:\Self\Data Science\Clustering\Assignment\EastWestAirlines.csv")
```

```python
data
```

```python
data.info()
```

```python
data=data.drop(['ID#'],axis=1)
data
```

```python
# Normalize heterogenous numerical data using z-score (x-mean/std) or custom defined function 
# Normalization function - here custom defined
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
# Normalized data frame (considering the numerical part of data)
data2_norm = norm_func(data)
data2_norm
```

```python
#  create Dendrograms
plt.figure(figsize=(20,10))
dendograms=sch.dendrogram(sch.linkage(data2_norm,'complete'))
```

```python
# create Clusters(y)
hclusters=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(data2_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
#Adding clusters to dataset
data['clustersid']=hclusters.labels_
data
```

```python
data.groupby('clustersid').agg(['mean']).reset_index()
```

```python
#plot Cluster
plt.figure(figsize=(10,7))
plt.scatter(data['clustersid'],data['Balance'], c=hclusters.labels_)
```
